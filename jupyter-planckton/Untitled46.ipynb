```python
from qiskit import QuantumCircuit, Aer, transpile, assemble

# Create the quantum circuit with 3 qubits and 3 classical bits
circuit = QuantumCircuit(3, 3)

# Prepare the state to be teleported
circuit.h(1)
circuit.cx(1, 2)

# Create entanglement between qubits 0 and 1
circuit.h(0)
circuit.cx(0, 1)

# Apply CNOT and Hadamard gates
circuit.cx(1, 2)
circuit.h(1)

# Measure qubits 0 and 1
circuit.measure([0, 1], [0, 1])

# Measure qubit 2 and store the result in classical bit 2
circuit.barrier()
circuit.cx(1, 2)
circuit.cz(0, 2)
circuit.measure(2, 2)

# Draw the circuit
print(circuit.draw())

# Simulate the circuit
simulator = Aer.get_backend('qasm_simulator')
compiled_circuit = transpile(circuit, simulator)
qobj = assemble(compiled_circuit)
result = simulator.run(qobj).result()
counts = result.get_counts()

print("Measurement outcomes:", counts)
```
