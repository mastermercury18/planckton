```python
def print_pascal_triangle(n):
    triangle = [[1]]
    for i in range(1, n):
        row = [1]
        for j in range(1, i):
            row.append(triangle[i-1][j-1] + triangle[i-1][j])
        row.append(1)
        triangle.append(row)
    
    # Calculate the width needed for the base of the triangle
    base_width = len(' '.join(map(str, triangle[-1])))
    
    for row in triangle:
        # Join all numbers in the row to a single string with space separation
        print_row = ' '.join(map(str, row))
        # Print each row centered relative to the base width
        print(print_row.center(base_width))
        
# Number of rows
n = 6
print_pascal_triangle(n)
```
