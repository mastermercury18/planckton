```python
def generate_pascal_triangle(rows):
    triangle = []
    for i in range(rows):
        row = [1]
        if triangle:
            last_row = triangle[-1]
            row.extend([sum(pair) for pair in zip(last_row, last_row[1:])])
            row.append(1)
        triangle.append(row)

    # Print the Pascal Triangle
    for row in triangle:
        print(" ".join(map(str, row)))

generate_pascal_triangle(6)
```

```python
def prime_factors(number):
    factor = 2
    while factor <= number:
        if number % factor == 0:
            print(factor)
            number = number // factor
        else:
            factor += 1

# Test the function with a number
number = 95  # You can change this number to find prime factors for a different number
print(f"Prime factors of {number}:")
prime_factors(number)
```

```python
def is_armstrong_number(number):
    # Calculate the number of digits
    num_digits = len(str(number))
    
    # Calculate the sum of digits raised to the num_digits power
    sum_of_digits = sum([int(digit) ** num_digits for digit in str(number)])
    
    # Check if the number is an Armstrong number
    if sum_of_digits == number:
        return True
    else:
        return False

# Test the function with a number
number = 24  # You can change this number to test with different numbers
if is_armstrong_number(number):
    print(f"{number} is an Armstrong number.")
else:
    print(f"{number} is not an Armstrong number.")
```

```python
def generate_armstrong_numbers():
    armstrong_numbers = []
    for number in range(1, 1000):
        num_digits = len(str(number))
        sum_of_digits = sum([int(digit) ** num_digits for digit in str(number)])
        if sum_of_digits == number:
            armstrong_numbers.append(number)
    
    return armstrong_numbers

# Generate Armstrong numbers up to 1000
armstrong_list = generate_armstrong_numbers()

# Print the Armstrong numbers and the total count
print("Armstrong numbers up to 1000:")
for armstrong_number in armstrong_list:
    print(armstrong_number)
print(f"Total Armstrong numbers up to 1000: {len(armstrong_list)}")
```
